```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
```

```python
url = "https://bimamedia-gurusiana.ap-south-1.linodeobjects.com/099fe6b0b444c23836c4a5d07346082b/2021/04/20/l-img20210420015823jpg20210420005933.jpeg"

image = io.imread(url)
image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.show()
```

```python
tinggi = image.shape[0]
lebar = image.shape[1]

image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis ditengah gambar dengan panjang tertentu

for y in range (95, 250):
  for x in range(160, 173):
    image[x,y] = [0,255,0]

for y in range (95, 250):
  for x in range(175, 185):
    image[x,y] = [203,192,255]

plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.show()
```
